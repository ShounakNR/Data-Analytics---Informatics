## Pass vs Run plays by down

```python
df.groupby('down')['encoded_plays'].value_counts().unstack(level=-1).plot(kind='bar',figsize=(15,7), fontsize= 14, rot = 0)
plt.xlabel('Down', fontsize=15)
plt.ylabel('Count',fontsize = 15)
plt.title('Plays by Down Season 14-18', fontsize = 20)
plt.legend(fontsize = 14)
```

## Pass vs run plays by down and distance to first down

```python
ax = df.groupby(['down','first_down_dis'])['play_type'].value_counts().unstack(level=-1).plot(kind='bar',figsize=(15,8),fontsize= 14, rot = 90)
plt.xlabel('(Down,1st_down_dis)', fontsize=15)
plt.ylabel('Count', fontsize = 15)
plt.title('Plays by Down & dist_to_1st_down Season(14-18)', fontsize = 20)
plt.text(0.5,0.6,'Dist_to_go goups: \n1:<3 yds\n2: 4-9 yds\n3:10-15 yds\n4:15+yds',size = 14,transform=ax.transAxes)
```

## Table summarizing league wide tendencies

```python
ydsTable = df.groupby(['down','first_down_dis'])['play_type'].value_counts().unstack(level=-1)
ydsTable['pass_call_percent'] = ydsTable['pass']/(ydsTable['pass']+ydsTable['run'])
ydsTable['run_call_percent'] = ydsTable['run']/(ydsTable['pass']+ydsTable['run'])

ydsTable = ydsTable.reset_index()
ydsTable.columns.name = None
ydsTable['yds_range']=ydsTable['first_down_dis'].apply(lambda x:'<3' if x == 1 else('4-9' if x == 2 else('10-15' if x == 3 else '20+')))

ydsTable =ydsTable[['down','yds_range','pass_call_percent','run_call_percent']]
ydsTable=ydsTable.set_index(['down','yds_range'])

format_dict = {'pass_call_percent': '{:.2%}','run_call_percent': '{:.2%}'}

ydsTable[['pass_call_percent','run_call_percent']].style.format(format_dict)
```

## Table summarizing team wise tendencies

```python
ydsTable = df.groupby(['posteam','down','first_down_dis'])['play_type'].value_counts().unstack(level=-1)
ydsTable['pass_call_percent'] = ydsTable['pass']/(ydsTable['pass']+ydsTable['run'])
ydsTable['run_call_percent'] = ydsTable['run']/(ydsTable['pass']+ydsTable['run'])

ydsTable = ydsTable.reset_index()
ydsTable.columns.name = None
ydsTable['yds_range']=ydsTable['first_down_dis'].apply(lambda x:'<3' if x == 1 else('4-9' if x == 2 else('10-15' if x == 3 else '20+')))

ydsTable =ydsTable[['posteam','down','yds_range','pass_call_percent','run_call_percent']]
ydsTable=ydsTable.set_index(['posteam','down','yds_range'])

format_dict = {'pass_call_percent': '{:.2%}','run_call_percent': '{:.2%}'}

ydsTable[['pass_call_percent','run_call_percent']].style.format(format_dict)
print("Teamwise play percentage")
```
