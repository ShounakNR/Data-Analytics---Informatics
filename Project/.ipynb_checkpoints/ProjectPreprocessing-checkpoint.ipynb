```python
play_df = pd.read_csv('combined.csv')
```

```python
play_df.to_csv( "combined.csv", index=False, encoding='utf-8-sig')
```

```python
play_df = play_df.loc[:, ~play_df.columns.str.contains('^Unnamed')]
```

```python
play_df=play_df.sort_values(by='GameId', ascending=False, na_position='first')
```

```python
col_counter=['IsIncomplete','IsInterception','IsFumble','IsPenalty','IsSack']
for col in col_counter:
    play_df.drop(play_df.loc[play_df[col]==1].index, inplace=True)
```

```python
cols = ['IsIncomplete','PassType', 'IsSack', 'IsChallenge',
       'IsChallengeReversed', 'Challenger', 'IsMeasurement', 'IsInterception',
       'IsFumble', 'IsPenalty','RushDirection', 'YardLineFixed',
       'YardLineDirection', 'IsPenaltyAccepted', 'PenaltyTeam', 'IsNoPlay',
       'PenaltyType', 'PenaltyYards','SeasonYear']

play_df=play_df.drop(cols, axis=1)
```
