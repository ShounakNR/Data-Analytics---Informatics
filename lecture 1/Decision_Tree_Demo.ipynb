```python
#import module of datasets from sklearn
from sklearn import datasets
```

```python
#load dataset of interest, for other datasets simply -> datasets.load_nameofdata()
breast = datasets.load_breast_cancer()
```

```python
#see what data type is returned from loaded dataset
type(breast)
```

```python
#data is loaded in two parts, data (which are attributes to train on) and target, assign these to vars
x,y = breast.data, breast.target
```

```python
#import tree module which we will access basic decision tree from
from sklearn import tree
```

```python
#Create the classfier structure with specified parameters (max depth is for how many levels will be made)
tree_clf = tree.DecisionTreeClassifier(max_depth=1)
tree_clf.fit(x,y)
#Fit the classifier to your data and target values
```

```python
#below gives a plot of your created tree
tree.plot_tree(tree_clf)
```

```python
#below is just an example of increasing max depth by 1 
tree_clf = tree.DecisionTreeClassifier(max_depth=2)
tree_clf.fit(x,y)
tree.plot_tree(tree_clf)
```

```python
#below are what the values are:
#value = ['# of malignant', '# of benign']
```

```python
#below changes the split criterion from gini to entropy
tree_clf = tree.DecisionTreeClassifier(max_depth=2,criterion='entropy')
tree_clf.fit(x,y)
tree.plot_tree(tree_clf)
```
